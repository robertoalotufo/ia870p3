# Area open

## Description
iaareaopen removes any grain (i.e., connected component) with area less than a of a binary image f. The connectivity is given by the structuring element Bc. This operator is generalized to gray-scale images by applying the binary operator successively on slices of f taken from higher threshold levels to lower threshold levels. For a faster implementation use iamxtareaopen

```python
import numpy as np
import ia870 as mm
from PIL import Image
```

## Numeric binary example

```python
f = np.array([
    [1, 1, 0, 0, 0, 0, 1],
    [1, 0, 1, 1, 1, 0, 1],
    [0, 0, 0, 0, 1, 0, 0]],bool)
y = mm.iaareaopen(f, 4, mm.iasecross())
print(y * 1)
```

## Numeric gray-level example

```python
f = np.array([
    [10,   11,   0,    0,   0,   0,  20],
    [10,    0,   5,    8,   9,   0,  15],
    [10,    0,   0,    0,  10,   0,   0]])
y = mm.iaareaopen(f, 4, mm.iasecross())
print(y)
```

## Binary image example

```python
a_pil = Image.open('data/form-1.tif')
a_pil
```

```python
a = np.array(a_pil) > 0
b = mm.iaareaopen(a,500)
Image.fromarray(b.astype(np.uint8)*255)
```

## Gray-level image example
This example uses gray scale area open.

```python
a_pil = Image.open('data/bloodcells.tif').convert('L') # must read as grayscale
a_pil
```

```python
a = np.array(a_pil)
b = mm.iaareaopen(a,500,mm.iasecross())
Image.fromarray(b)
```
